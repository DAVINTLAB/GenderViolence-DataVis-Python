```python
# Função para pegar a primeira palavra antes de ">", "." ou ";"
def get_first_word(text):
    if isinstance(text, str):
        return re.split(r'[>.;]', text)[0]
    return None

df['primeira_palavra'] = df['Violações'].apply(get_first_word)

primeiras_palavras_distintas = df['primeira_palavra'].dropna().value_counts()
print(primeiras_palavras_distintas)
```

```python
# Função para pegar o texto após a primeira palavra
def get_following_text(text):
    if isinstance(text, str) and text.startswith("INTEGRIDADE"):
        return re.split(r'[>.;]', text, 1)[1] if len(re.split(r'[>.;]', text)) > 1 else None
    return None

# Aplicando a função para pegar os valores após "INTEGRIDADE"
df['seguinte_integridade'] = df['Violações'].apply(get_following_text)

valores_seguinte_integridade = df['seguinte_integridade'].dropna().value_counts()
print(valores_seguinte_integridade)
```

```python
# Função para pegar o texto após a primeira palavra
def get_following_text(text):
    if isinstance(text, str) and text.startswith("LIBERDADE"):
        return re.split(r'[>.;]', text, 1)[1] if len(re.split(r'[>.;]', text)) > 1 else None
    return None

# Aplicando a função para pegar os valores após "INTEGRIDADE"
df['seguinte_integridade'] = df['Violações'].apply(get_following_text)
# Formatar cada linha para começar com "INTEGRIDADE>"
formatted_output = "\n".join([f"LIBERDADE>{index} {value}" for index, value in valores_seguinte_integridade.items()])

print(formatted_output)
```

```python
def get_full_text(text):
    if isinstance(text, str) and (text.startswith("LIBERDADE") or text.startswith("INTEGRIDADE")):
        return text  # Retorna a string completa
    return None

# Aplicando a função para capturar as strings inteiras que começam com "LIBERDADE" ou "INTEGRIDADE"
df['liberdade_integridade_completa'] = df['Violações'].apply(get_full_text)

# Exibindo os valores que começam com "LIBERDADE" ou "INTEGRIDADE"
valores_liberdade_integridade = df['liberdade_integridade_completa'].dropna().value_counts()
print(valores_liberdade_integridade)
```
