## Separador de palavras

```python
palavras_Canal_atend = {}

for texto in df['Canal de atendimento'].dropna():
    # Divide o texto em palavras
    palavras = str(texto).split()
    for palavra in palavras:
        if palavra not in palavras_Canal_atend:
            palavras_Canal_atend[palavra] = None  #depois substituir esse None pelas palavras traduzidas

print(palavras_Canal_atend)
```

```python
# Criação do dicionário de traduções
palavras_Canal_atend = {
    'ATENDIMENTO': 'service',
    'TELEFÔNICO': 'telephone',
    'APLICATIVOS/ONLINE': 'apps/online',
    'WEB/ONLINE': 'web/online',
    'CHAT': 'chat',
    'E-MAIL': 'email',
    'VÍDEO/OUTROS': 'video/others',
    'CORRESPONDÊNCIA/PRESENCIAL': 'correspondence/in-person',
    'E-MAIL:': 'email:',
    'LIGUE180@MDH.GOV.BR': 'ligue180@mdh.gov.br',
    'DENUNCIA': 'complaint',
    '-': '-',
    'TELEGRAM': 'telegram'
}

# Função para substituir palavras em cada célula
def traduzir_canal(texto, dicionario):
    if isinstance(texto, str): 
        palavras = texto.split()  # Dividir o texto em palavras
        palavras_traduzidas = [dicionario.get(palavra, palavra) for palavra in palavras]  # Traduzir cada palavra
        return ' '.join(palavras_traduzidas)  # Juntar novamente as palavras traduzidas
    return texto

df_traduzido = df.copy()
df_traduzido['Canal de atendimento'] = df['Canal de atendimento'].apply(traduzir_canal, dicionario=palavras_Canal_atend)
print(df_traduzido.head())
```
