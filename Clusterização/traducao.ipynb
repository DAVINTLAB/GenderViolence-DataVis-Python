```python
# Função para substituir palavras em cada célula
def traduzir_canal(texto, dicionario):
    if isinstance(texto, str): 
        palavras = texto.split()  # Dividir o texto em palavras
        if all(palavra.lower() in dicionario.values() for palavra in palavras):
            return texto  # verifica se já tá traduzido
        palavras_traduzidas = [dicionario.get(palavra, palavra) for palavra in palavras]  # Traduzir cada palavra
        return ' '.join(palavras_traduzidas)  # Junta palavras traduzidas
    return texto
```

```python
# Dicionários de tradução (exemplos)
dicionarios_traducao = {
    'Canal de atendimento': palavras_Canal_atend,
    'Denúncia emergencial': palavras_denuncia_emerg,
    'Denunciante': palavras_denunciante,
    'Cenário da violação': palavras_cenario_viola

}

# Loop que faz a tradução em todas as colunas de uma vez
for coluna, dicionario in dicionarios_traducao.items():
    df_traduzido[coluna] = df[coluna].apply(traduzir_canal, dicionario=dicionario)


print(df_traduzido.head())
```
