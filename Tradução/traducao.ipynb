# Aplica a tradução

```python
def TraduzDataset(df, dicionarioDeTraducao):
    """
    Traduz o dataset com base no dicionário de traduções fornecido.
    
    Args:
    df (pd.DataFrame): DataFrame com os dados originais.
    translations (dict): Dicionário contendo as traduções de cada coluna.
    
    Returns:
    pd.DataFrame: DataFrame com os valores traduzidos.
    """
    dfCopy = df.copy()
    # Itera sobre cada coluna e verifica se há uma tradução definida no dicionário
    for coluna in dfCopy.columns:
        if coluna in dicionarioDeTraducao:
            # Aplica o mapeamento de tradução na coluna
            dfCopy[coluna] = dfCopy[coluna].map(dicionarioDeTraducao[coluna])
    return dfCopy

def TraduzColunaDeViolacao(coluna):
    # Função interna para traduzir uma string de tipos de violência
    def TraduzTexto(stringDaViolacao):
        for pt, en in dicionarioParaViolacao.items():
            # Substitui cada ocorrência de pt (português) pelo equivalente en (inglês)
            stringDaViolacao = re.sub(rf"\b{pt}\b", en, stringDaViolacao)
        return stringDaViolacao
    
    # Aplicar a função de tradução em cada elemento da coluna sem alterar o tipo
    return coluna.apply(TraduzTexto)
```
